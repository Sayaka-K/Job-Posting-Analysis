```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
```

```python
df= pd.read_csv('Data_analyst_cleaned.csv')
df.columns
```

```python
df = df[df['Salary_Estimate']!= '-1']
Salary= df['Salary_Estimate'].apply(lambda x: x.split('(')[0])
DeletedKd= Salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python

df['Min_salary'] = DeletedKd.apply(lambda x: int(x.split('-')[0]))
df['min_salary'] =df['Min_salary']
```

```python
df['Max_salary'] = DeletedKd.apply(lambda x: int(x.split('-')[1]))
df['max_salary'] = df['Max_salary']
```

```python

df['avg_salary'] = ((df.Min_salary+df.Max_salary)/2)
df['Avg_salary']= df['avg_salary']
```

```python
df['Job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df.Job_state.value_counts()
df['Job_state']
```

```python
df['Job_city'] = df['Location'].apply(lambda x: x.split(',')[0])

df['Job_city']
```

```python
def title_simplifier(title):
    if 'engineer' in title.lower():
        return 'Data Engineer'
   
    elif ' business analyst' in title.lower():
        return 'Business analyst'
    elif 'scientist' in title.lower():
        return 'Data Scientist'
    elif 'manager'in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'Director'
    elif 'data analyst' in title.lower():
        return 'data analyst'
    elif 'quantitative' in title.lower():
        return 'quantitative analyst'
    
    elif 'operations' in title.lower():
        return 'operations analyst'
    elif 'marketing' in title.lower():
        return 'marketing analyst'
#     else :
#         return 'Na'
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'Sr.' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'Junior' in title.lower() or 'junior' in title.lower():
        return 'junior'
#     else :
#         return 'na'
df['job_simp'] = df['Job_Title'].apply(title_simplifier) 
```

```python
df['headquarters_city'] = df['Headquarters'].apply(lambda x: x.split(',')[0])
df['headquarters_city']
df['Same_state_hq_jl'] = df.apply(lambda x:1 if x.Location == x.Headquarters else 0 , axis=1)
df['Same_state_hq_jl']
```

```python
df['Age']= df.Founded.apply(lambda x:x if x<1 else 2020-x)
df['Age']
```

```python
df['python_jd']= df['Job_Description'].apply(lambda x: 1 if 'python' in x.lower() else 0 )
df.python_jd.value_counts()
```

```python
df['Sql']= df['Job_Description'].apply(lambda x: 1 if 'sql'  in x.lower() or 'SQL' in x.lower() else 0 )
df.Sql.value_counts()
```

```python
df['aws']= df['Job_Description'].apply(lambda x: 1 if 'aws'  in x.lower() or 'amazon' in x.lower()  else 0 )
df.aws.value_counts()
```

```python
df['visualization']= df['Job_Description'].apply(lambda x: 1 if 'visualize '  in x.lower() or 'tableau' in x.lower() or 'visualization' in x.lower()  else 0 )
df.visualization.value_counts()
```

```python
df['Project_Management']= df['Job_Description'].apply(lambda x: 1 if 'Project Management'  in x.lower() or 'team work' in x.lower() or 'project' in x.lower()  else 0 )
df.Project_Management.value_counts()
```

```python
df['Database_Management']= df['Job_Description'].apply(lambda x: 1 if 'databasemanagement'  in x.lower() or 'dbs' in x.lower() or 'database' in x.lower()  else 0 )
df.Database_Management.value_counts()
```

```python
df['Excel']= df['Job_Description'].apply(lambda x: 1 if 'excel'  in x.lower() or 'Ms' in x.lower() else 0 )
df.Excel.value_counts()
df['Excel']
```
